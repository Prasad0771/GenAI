```python
#SimpleDocTemplate()	Creates and manages the PDF document
#getSampleStyleSheet()	Provides ready-made text styles
#Paragraph()	Adds formatted text to PDF
from reportlab.platypus import SimpleDocTemplate, Paragraph
from reportlab.lib.styles import getSampleStyleSheet
```

```python
### Tree of Thoughts Non linear
from openai import OpenAI

client = OpenAI(api_key=api_key1)

prompt = """
You are a Senior Data Scientist.

Problem:

what is Supervised Machine Learning in 20 pages of PDF?
Answer:
Explain different types of it?

For each type:
- Introduction in 5 pages
- Flow Diagram of each type
- Tables
- Generate text

"""

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[{"role":"user","content":prompt}]
)

#print(response.choices[0].message.content)
txt = response.choices[0].message.content
pdf = SimpleDocTemplate("sml.pdf")
Styles = getSampleStyleSheet()
pdf.build([Paragraph(txt,Styles["BodyText"])])
print("The PDF is created.")
```
