```python
### Creating the word2Vec model of words of the single PDF
# pip install pypdf gensim nltk

from pypdf import PdfReader
from gensim.models import Word2Vec
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')

## Read PDF

reader = PdfReader("F:\Bharat DS & DE 23 July 2025\Gen AI\my_Gen AI\data_science_01_david_kim.pdf")

text = ""
for page in reader.pages:
    if page.extract_text():
        text += page.extract_text()

# Convert text into sentences
sentences = nltk.sent_tokenize(text)

# Convert each sentence into words
tokenized_sentences = [nltk.word_tokenize(sentence.lower())
                      for sentence in sentences
 ]

print(tokenized_sentences[:3])
model = Word2Vec(sentences = tokenized_sentences,
                 vector_size=50,
                 window=5,
                 min_count=1)

model.save("pdf_word2vec.model")
model.wv["data"]
model.wv.most_similar("data")
```
