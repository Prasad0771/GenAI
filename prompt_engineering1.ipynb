```python
from openai import OpenAI
client=OpenAI(api_key=api_key1)
while True:
  user_input = input("You: ")
  if user_input.lower() == 'exit':
    break
  else:
    prompt = '''Please make sure that all responses are in 5 bullet points
                for each point you have introduction,history, examples'''
    response = client.chat.completions.create(
        model = 'gpt-4.1-mini',
        messages =[   {"role":"system",
                      "content":"You are a helpful AI assistent"},
                     {"role":"user", "content": user_input+prompt}
                     ]
    )
    print(response.choices[0].message.content)
```

```python
### zero shot prompting
from openai import OpenAI
client = OpenAI(api_key=api_key1)

user_input = "Explain Machine Learning"

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {
            "role": "system",
            "content": "You are a helpful AI assistant"
        },
        {
            "role": "user",
            "content": user_input
        }
    ]
)

print(response.choices[0].message.content)
```

```python
### Few shot prompting
from openai import OpenAI

client = OpenAI(api_key=api_key1)

user_input = "What is Data Engineering?"

prompt = """
Use the below examples as a template for answers.

Example 1:

Question: What is Machine Learning?

Answer:
- Introduction: Machine Learning is a branch of Artificial Intelligence that enables systems to learn from data.
- History: The concept originated in the 1950s and gained popularity with increased computing power.
- Example: Predicting house prices using historical sales data.

Example 2:

Question: What is Data Science?

Answer:
- Introduction: Data Science is the field of extracting insights from structured and unstructured data.
- History: It evolved from statistics, mathematics, and computer science.
- Example: Customer churn prediction using historical customer behavior.

Example 3:

Question: What is Cloud Computing?

Answer:
- Introduction: Cloud Computing provides on-demand access to computing resources over the internet.
- History: Cloud services became mainstream in the mid-2000s with providers like AWS.
- Example: Hosting a web application on AWS EC2.

Now answer the following question in the same format.

Question: """ + user_input

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ]
)

print(response.choices[0].message.content)
```

```python
### Chain of Thoughts(COT) ex1 - Linear or sequential actions
from openai import OpenAI

client = OpenAI(api_key=api_key1)

prompt = """
Solve the problem step by step.

Question:
A company has 500 employees.
20% work in Data Engineering.
How many employees work in Data Engineering?

Show your reasoning before the final answer.
"""

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role": "user", "content": prompt}
    ]
)

print(response.choices[0].message.content)
```

```python
### Chain of Thoughts(COT) Ex2 Linear or sequential actions
from openai import OpenAI

client = OpenAI(api_key=api_key1)

prompt = """
Solve the problem step by step.

Question:

A retail company sold 15,000 products in January.

- 40% were Electronics
- 35% were Clothing
- Remaining were Groceries

Out of Electronics:
- 10% were returned

Out of Clothing:
- 5% were returned

Calculate:

1. Number of Electronics sold
2. Number of Clothing sold
3. Number of Groceries sold
4. Total returned products
5. Final products retained after returns

Show all calculations before the final answer.
"""

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role":"user","content":prompt}
    ]
)

print(response.choices[0].message.content)
```

```python
### Tree of Thoughts Non linear
from openai import OpenAI

client = OpenAI(api_key=api_key1)

prompt = """
You are a Senior Data Architect.

Problem:
A retail company processes 50 TB of data daily.

Generate 3 possible architectures.

For each architecture:
- Components
- Advantages
- Disadvantages

Compare all architectures and recommend the best one.
"""

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role":"user","content":prompt}
    ]
)

print(response.choices[0].message.content)
```

```python
### Tree of Thoughts Non linear
from openai import OpenAI

client = OpenAI(api_key=api_key1)

prompt = """
You are a Senior Data Scientist.

Problem:

A retail company has observed that 20% of customers
stop purchasing within 6 months.

Explore multiple approaches to predict customer churn.

Approach 1:
Logistic Regression

Approach 2:
Random Forest

Approach 3:
XGBoost

For each approach:
- Required features
- Advantages
- Disadvantages
- Expected performance

Compare all approaches and recommend the best solution for production.
"""

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[{"role":"user","content":prompt}]
)

print(response.choices[0].message.content)
```

```python
### Tree of Thoughts - Non linear
from openai import OpenAI

client = OpenAI(api_key=api_key1)

prompt = """
Customer Churn Increased
│
├── Option 1
│   Discount Campaign
│
├── Option 2
│   Loyalty Rewards Program
│
├── Option 3
│   Personalized Product Recommendations
│
└── Compare
     ├── Cost
     ├── ROI
     ├── Implementation Time
     └── Customer Impact

Best Strategy Selected
"""

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[{"role":"user","content":prompt}]
)

print(response.choices[0].message.content)
```
