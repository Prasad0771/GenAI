```python
### Creating the word2Vec model of texts of the all PDFs
from pypdf import PdfReader
from gensim.models import Word2Vec
import nltk
import glob

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
final_text = ""
#reading files from folder
for i in glob.glob(r"F:\Bharat DS & DE 23 July 2025\Gen AI\resumes\resumes/*.pdf"):



    ## Read PDF

    reader = PdfReader(i)

    text = ""
    for page in reader.pages:
        if page.extract_text():
            text += page.extract_text()
    final_text = final_text + text
    
```

```python
for i in glob.glob(r"F:\Bharat DS & DE 23 July 2025\Gen AI\resumes\resumes/*.pdf"):
    print(i)
```

```python
#Convert text into sentences
sentences = nltk.sent_tokenize(final_text)
# Convert each sentence into words
tokenized_sentences = [nltk.word_tokenize(sentence.lower())
                      for sentence in sentences
]

print(tokenized_sentences[:3])
```

```python
models = Word2Vec(
    sentences=tokenized_sentences,
    vector_size=100,
    window=5,
    min_count=1,
    #workers=4
)
models.save("folder_multiple pdfs_word2vec.model")
models.wv["data"]
models.wv.most_similar("data")
```
